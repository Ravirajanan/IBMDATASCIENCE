### In .open command use the photo path here i have the picture in same path.

```python

# load and show an image with Pillow
from PIL import Image
# Open the image form working directory
image = Image.open('picture_small.jpg')
# summarize some details about the image
print(image.format)
print(image.size)
print(image.mode)
# show the image
#load_image.show()
```

```python
# load and display an image with Matplotlib
from matplotlib import image
from matplotlib import pyplot
# load image as pixel array
image = image.imread('picture_small.jpg')
# summarize shape of the pixel array
print(image.dtype)
print(image.shape)
# display the array of pixels as an image
pyplot.imshow(image)
pyplot.show()
```

```python
from PIL import Image
from numpy import asarray
# load the image
image = Image.open('picture_small.jpg')
# convert image to numpy array
data = asarray(image)
print(type(data))
# summarize shape
print(data.shape)

# create Pillow image
image2 = Image.fromarray(data)
print(type(image2))

# summarize image details
print(image2.mode)
print(image2.size)
print(data)
```

```python
import numpy as np
load_img_rz = np.array(Image.open('picture_small.jpg').resize((300,300)))
Image.fromarray(load_img_rz).save('r_kolala.jpeg')
print("After resizing:",load_img_rz.shape)
```
