### DATA
At first the data is collected from google news website which provides us all the covid-19 cases around the world.

By using the BeautifulSoup the data is extracted from the 'https://news.google.com/covid19/map?hl=en-IN&gl=IN&ceid=IN%3Aen&mid=%2Fm%2F03rk0' website

```python
source = requests.get('https://news.google.com/covid19/map?hl=en-IN&gl=IN&ceid=IN%3Aen&mid=%2Fm%2F03rk0').text

soup = BeautifulSoup(source, 'lxml')
```

```python
table = soup.find('table')
```

The data is transfered in to a Data Frame using Pandas

```python
column_names = ['Location','Confirmed_cases','cases_per_million','Recoverd','Deaths'] 

# instantiate the dataframe
data = pd.DataFrame(columns=column_names)
```

```python
first = 1
cc='o'
cpm='1'
re='2'
de='3'

for row in table.find_all('tr'):
    location = row.th.text
    lis = 0
    for eachdata in row.find_all('td'):
        if lis == 0:
            cc = (eachdata.text)
            cc = cc.replace(',', '')
            cc=int(cc)
            lis = lis+1
        elif lis == 1:
            cpm = (eachdata.text)
            if cpm != '—':
                cpm = cpm.replace(',', '')
                if cpm != "No data":
                    if cpm:
                        cpm=int((cpm))
                else:
                    cpm = 0
            
            lis = lis+1
        elif lis == 2:
            re = eachdata.text
            lis = lis+1
        elif lis == 3:
            de = (eachdata.text)
            if de != '—':
                de = de.replace(',','')
                if de != "No data":
                    if de:
                        de=int((de))
                else:
                    de = 0
            lis = lis +1
    data = data.append({'Location':location,'Confirmed_cases':cc,'cases_per_million':cpm,'Recoverd':re,'Deaths':de},ignore_index=True)
data.head()
```

```python
data_india_cases = df_india[df_india['new_cases'] != 0]
data_india_cases
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

Seperating the confirmed cases data to a new data frame.

```python
ml_data = data_india_cases[['date','new_cases']]
ml_data.set_index('date', inplace = True)
ml_data.head()
```

```python
ml_data.index.to_period('D')
train, test = ml_data.iloc[:len(ml_data)-5], ml_data.iloc[len(ml_data)-5:]
```

### Applying the ARIME (AutoRegressive Integrated Moving Average) Time serires analysis model on data.

Training the model

```python
from pmdarima import auto_arima
model = auto_arima(ml_data, trace=True, error_action='ignore',D=1,h=365, m=7,suppress_warnings=True)
model.fit(train)
result = model.fit(train)
result.summary()
```

Predicting the values for the test data

```python

forecast = model.predict(n_periods=len(test))
forecast = pd.DataFrame(forecast,index = test.index,columns=['Prediction'])

#plot the predictions for validation set
plt.plot(ml_data, label='Train')
plt.plot(test, label='Valid')
plt.plot(forecast, label='Prediction')
plt.show()
```

Calculating RMS

```python
from math import sqrt
from sklearn.metrics import mean_squared_error

rms = sqrt(mean_squared_error(test,forecast))
print(rms)
```
